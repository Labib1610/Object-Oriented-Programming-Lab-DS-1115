```python
def element_frequency(lst):
    dictionary = {}
    for i in lst:
        a = lst.count(i)
        dictionary[i] = a
    return dictionary
print(element_frequency([2,4,4,1,1,3,9]))
```

```python
def running_total(lst):
    lst2 =[]
    count = 0
    for i in range(len(lst)):
        count = count+lst[i]
        lst2.append(count)
    return lst2
print(running_total([1,2,3,4,5]))
```

```python
def remove_duplicate_from_string(input_str):
    lst = [i for i in input_str]
    lst2 = []
    for i in lst:
        if i not in lst2:
            lst2.append(i)
    string = ''
    for i in lst2:
        string = string+i
    return string
print(remove_duplicate_from_string('programming'))
```

```python
def are_anagrams(word1,word2):
    lst = [i for i in word1]
    len_of_lst = len(lst)
    count = 0
    for i in word2:
        if i in lst:
            count = count+1
            lst.remove(i)
    if count == len_of_lst:
        return True
    else:
        return False
print(are_anagrams('silent','listen'))
```
